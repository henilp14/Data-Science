```python
import pandas as pd
import numpy as np

# Sample data
np.random.seed(42)
n = 100
experience = np.random.randint(1, 21, size=n)  # Years of experience
salary = 40000 + experience * 2000 + np.random.normal(0, 5000, n)  # Salary based on experience

# Create DataFrame
df = pd.DataFrame({'Experience': experience, 'Salary': salary})
print (df)

# Calculate Pearson correlation coefficient
correlation = df['Experience'].corr(df['Salary'])
print(f"Pearson Correlation between Experience and Salary: {correlation}")

# Calculate Spearman rank correlation
spearman_corr = df['Experience'].corr(df['Salary'], method='spearman')
print(f"Spearman Rank Correlation between Experience and Salary: {spearman_corr}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to plot correlation
def plot_correlation(ax, x, y, title):
    ax.scatter(x, y)
    ax.axhline(0, color='red', linestyle='--', label='Zero Correlation Line')
    ax.set_title(title)
    ax.legend()

# Generate data
np.random.seed(0)
x = np.random.rand(100)

# Perfect positive correlation
y_perfect_pos = x
# Moderate positive correlation
y_moderate_pos = x + np.random.normal(0, 0.2, 100)
# No correlation
y_no_corr = np.random.rand(100)
# Moderate negative correlation
y_moderate_neg = -x + np.random.normal(0, 0.2, 100)
# Perfect negative correlation
y_perfect_neg = -x

fig, axs = plt.subplots(3, 2, figsize=(12, 12))
fig.delaxes(axs[2, 1])  # Delete the extra subplot

plot_correlation(axs[0, 0], x, y_perfect_pos, "1: Perfect positive correlation")
plot_correlation(axs[0, 1], x, y_moderate_pos, "0.5: Moderate positive correlation")
plot_correlation(axs[1, 0], x, y_no_corr, "0: No correlation")
plot_correlation(axs[1, 1], x, y_moderate_neg, "-0.5: Moderate negative correlation")
plot_correlation(axs[2, 0], x, y_perfect_neg, "-1: Perfect negative correlation")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
np.random.seed(0)
x = np.linspace(-10, 10, 100)

# Define different correlations
y_perfect_pos = x  # Perfect positive correlation (1)
y_moderate_pos = 0.5 * x  # Moderate positive correlation (0.5)
y_no_corr = np.random.normal(0, 1, 100)  # No correlation (0)
y_moderate_neg = -0.5 * x  # Moderate negative correlation (-0.5)
y_perfect_neg = -x  # Perfect negative correlation (-1)

plt.figure(figsize=(10, 6))

# Plot each correlation
plt.plot(x, y_perfect_pos, label='1: Perfect positive correlation', linestyle='-', color='green')
plt.plot(x, y_moderate_pos, label='0.5: Moderate positive correlation', linestyle='--', color='blue')
plt.plot(x, y_no_corr, label='0: No correlation', linestyle=':', color='black')
plt.plot(x, y_moderate_neg, label='-0.5: Moderate negative correlation', linestyle='-.', color='orange')
plt.plot(x, y_perfect_neg, label='-1: Perfect negative correlation', linestyle='-', color='red')

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph Representing Different Correlation Coefficients')
plt.legend()

plt.show()
```
