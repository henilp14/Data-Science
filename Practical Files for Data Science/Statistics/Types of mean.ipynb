```python
import numpy as np
import pandas as pd
from scipy import stats
df= pd.read_csv("C:/Users/al6436/Downloads/vs.csv")
```

```python
# trimmed mean
data = df['Amount']
trimmed_mean = stats.trim_mean(data,0.1)
print(trimmed_mean)
```

```python
# Weight Mean
data = {
    'Category': ['Blouse', 'Bottom', 'Ethnic Dress', 'Kurta', 'Saree', 'Set', 'Top', 'Western Dress'],
    'Sum of Amount': [140888, 27804, 195256, 4959377, 1010471, 10507546, 1186199, 3148836],
}
data1 = pd.DataFrame(data)
data1['Weight'] = data1['Sum of Amount'] / data1['Sum of Amount'].sum()
print(data1)
weighted_mean = (data1['Sum of Amount'] * data1['Weight']).sum()
print("weighted_mean", weighted_mean)
```

```python
# Geometric Mean
data = pd.DataFrame({'Month': df['Month'], 'Amount': df['Amount']})
geometric_mean = stats.mstats.gmean(data['Amount'])
print("geometric_mean", geometric_mean)
```

```python
# Harmonic Mean 
monthly_harmonic_mean = df.groupby('Month')['Amount'].apply(stats.mstats.hmean)
print(monthly_harmonic_mean)
```

```python
# Moving Average (SMA and EMA )
df['Date'] = pd.to_datetime(df['Date'])

df['SMA_1'] = df['Amount'].rolling(window=1).mean()
df['EMA_1'] = df['Amount'].ewm(span=1, adjust=False).mean()

# Print the result
print(df['EMA_1'])
print(df['SMA_1'])
```

```python
import scipy.stats as stats

df = 7

alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, df)

print("Critical Value at 0.05 significance level:", critical_value)
```

```python
import numpy as np
import scipy.stats as stats
import pandas as pd

# Data for the contingency table (counts of Men and Women across categories)
data = {
    'Blouse': [6, 210],  # [Men, Women]
    'Bottom': [39, 26],
    'Ethnic Dress': [160, 85],
    'kurta': [154, 9297],
    'Saree': [271, 997],
    'Set': [4896, 6870],
    'Top': [45, 1995],
    'Western Dress': [3185, 405]
}

# Create a DataFrame
df = pd.DataFrame(data, index=['Men', 'Women'])

# Perform the Chi-Square Test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(df)

# Print the results
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequency Table:\n{expected}")
```
