```python
import streamlit as st
import pandas as pd
import os

# Set title
st.title("Student Marks Uploader & Excel Generator")

# User Inputs
name = st.text_input("Enter your Name")
college = st.text_input("Enter your College Name")
uploaded_file = st.file_uploader("Upload a CSV file with Student Marks", type=["csv"])

if st.button("Save Data"):
    if name and college and uploaded_file is not None:
        # Read uploaded CSV file
        marks_df = pd.read_csv(uploaded_file)
        
        # Create DataFrame with all information
        student_info = pd.DataFrame({
            "Name": [name] * len(marks_df),
            "College": [college] * len(marks_df)
        })
        
        final_df = pd.concat([student_info, marks_df], axis=1)

        # File name
        file_name = "student_data.xlsx"
        
        # Check if file exists and append data
        if os.path.exists(file_name):
            existing_data = pd.read_excel(file_name)
            final_df = pd.concat([existing_data, final_df], ignore_index=True)

        # Save to Excel
        final_df.to_excel(file_name, index=False)

        st.success("Data successfully saved to 'student_data.xlsx'!")
        st.dataframe(final_df)  # Show data in Streamlit
    else:
        st.error("Please fill in all fields and upload a file!")
```
